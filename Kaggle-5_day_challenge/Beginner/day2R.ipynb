```r
library(tidyverse)
```

```r
# load the dataset
data <- read_csv('../input/camera_dataset.csv')
```

```r
# summarize the data
summary(data)
head(data)
```

```r
# rename `Release date` to `ReleaseDate` before plotting
names(data)[names(data) == "Release date"] <- "ReleaseDate"

# plot the `Release date`
data %>%
    ggplot(aes(ReleaseDate, fill = cut(ReleaseDate, 100))) +
    geom_histogram(binwidth=0.4, show.legend=F) +
    scale_fill_discrete(h = c(180, 360), c = 150, l = 80) +
    ggtitle("Camera Count by Release Date")
```
