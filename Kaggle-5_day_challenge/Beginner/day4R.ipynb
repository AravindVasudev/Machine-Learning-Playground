```r
library(tidyverse)
```

```r
# read the dataset
data <- read_csv("../input/Health_AnimalBites.csv")

head(data)
```

```r
# plot bar-chart on SpeciesIDDesc
data %>%
    filter(!is.na(SpeciesIDDesc)) %>%
    ggplot(aes(SpeciesIDDesc)) +
    geom_bar() +
    ggtitle("Species Bar-chart")
```

```r
# plot bar-chart on GenderIDDesc
data %>%
    filter(!is.na(GenderIDDesc)) %>%
    ggplot(aes(GenderIDDesc)) +
    geom_bar() +
    ggtitle("Gender Bar-chart")
```
