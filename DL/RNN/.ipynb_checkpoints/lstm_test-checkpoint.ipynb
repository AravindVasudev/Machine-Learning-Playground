```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
# dataset
X = np.arange(100).reshape(1, 1, 100)
y = X + 1

X_test = np.arange(100, 200).reshape(1, 1, 100)
y_test = X_test + 1
```

```python
# ** model start **
model = Sequential()

model.add(LSTM(100, input_shape=(1, 100), return_sequences=True))
model.add(Dropout(0.2))

model.add(Dense(100))
# ** model end **

model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# train the model
model.fit(X, y, epochs=10000, batch_size=1, verbose=2, validation_data=(X_test, y_test))
```

```python
model.predict(X)
```

```python
model.predict(X_test)
```
