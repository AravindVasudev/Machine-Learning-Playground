```python
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
from skimage import io, color
from skimage import exposure

%matplotlib inline
```

```python
img = io.imread('./apple.jpg')

plt.imshow(img)
```

```python
img = color.rgb2gray(img)

plt.imshow(img, cmap=plt.cm.gray)
```

```python
img.shape
```

```python
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
```

```python
edges = scipy.signal.convolve2d(img, kernel, 'valid')
```

```python
plt.imshow(edges, cmap=plt.cm.gray)
```

```python
edges_equalized = exposure.equalize_adapthist(edges/np.max(np.abs(edges)), clip_limit=0.03)
```

```python
plt.imshow(edges_equalized, cmap=plt.cm.gray) 
```

```python
k2 = np.array([[0, 0, 0], [-1, 1, 0], [0, 0, 0]])
```

```python
e2 = scipy.signal.convolve2d(img, k2, 'valid')
```

```python
plt.imshow(e2, cmap=plt.cm.gray)
```

```python
edges_equalized2 = exposure.equalize_adapthist(e2/np.max(np.abs(e2)), clip_limit=0.03)
```

```python
plt.imshow(edges_equalized2, cmap=plt.cm.gray)
```

```python
e3 = scipy.signal.convolve2d(img, k2)
```

```python
plt.imshow(e3, cmap=plt.cm.gray)
```
